### Research Question 2  (what is the year that released highest number of movies ?)

```python
movies['release_year'].value_counts().reset_index().sort_values(by = 'index')
```

```python
movies['release_year'].value_counts().reset_index().max()
```

```python
# Continue to explore the data to address your additional research
#   questions. Add more headers as needed if you have more questions to
#   investigate.

plt.figure(figsize = (15,10));
x = movies['release_year'].value_counts().reset_index().sort_values(by = 'index')
plt.plot(x['index'],x['release_year']);
plt.xticks(np.arange(1950,2018,1),rotation = 90);
plt.xlabel('Year',fontsize = 15);
plt.ylabel('Number of Movies', fontsize  =15)
plt.title('movies production through time');
```

### There is increasing in movies production through years

### Research Question 3  (What are the most popular genres ?)

```python
movies_geners = pd.merge(movies,genres,on  = 'id')
movies_geners.drop(columns=['budget', 'revenue', 'original_title', 'director',
       'overview', 'release_date', 'vote_count', 'vote_average',
       'release_year', 'budget_adj', 'revenue_adj'])
```

```python
movies_geners['popularity'].max()
```

```python
plt.figure(figsize = (15,6));
sns.barplot(x = movies_geners.groupby('genres').max()['popularity'].index, 
            y = movies_geners.groupby('genres').max()['popularity'].sort_values(ascending = True));
plt.xlabel('Genres', fontsize = 12);
plt.ylabel('Popularity Rank', fontsize = 12);
plt.xticks(rotation = 90);
plt.yticks(np.arange(0,40,2))
plt.title('High popularity geners');
```

### It showes that  Tv-movies,Thriller,war,westem are the most popularity.
